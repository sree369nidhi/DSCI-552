```python


wholeData = newdataset.drop(['activity'], axis=1)
print("columns \t\t std")
print("-------------------------------")
print(wholeData.std())
```

```python
column = []
lowerbound = []
upperbound = []
std =[]
finaltable = {}
for i in wholeData:
    column.append(i)
    bootResult = bs.bootstrap(np.array(newdataset[i]), stat_func=bs_stats.std, alpha=0.1)
    lowerbound.append(bootResult.lower_bound)
    upperbound.append(bootResult.upper_bound)
    std.append(newdataset[i].std())
finaltable['columns'] = column
finaltable['std'] = std
finaltable['lowerbound'] = lowerbound
finaltable['upperbound'] = upperbound
pd.DataFrame.from_dict(finaltable,orient='columns')

```

## 2. ISLR 3.7.4

### (a) Linear Train

By adding more polynomial terms, the cubic regression can adapt more to the training data (The extra polynomial terms allow more degrees of freedom of the training data), thus it's likely to have a lower training RSS than the simple linear regression.

### (b) Linear Test

Linear regression will predict better on unseen data, meaning it will have a lower test RSS. The cubic model might overfit the training data, leading to a higher test RSS.

### (c) Not Linear Train

As the cubic regression better fits non-linear data, it's expected to have a lower training RSS.
    

### (d) Not Linear Testing

The test RSS would depend on the proximity of the true relationship f(x) to a linear one. If f(x) is closer to being linear, cubic regression might overfit, leading to a higher cubic RSS and lower linear RSS. Conversely, if f(x) is closer to being cubic, linear regression might underfit, resulting in a higher linear RSS and lower cubic RSS.
